```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
shapec = gpd.read_file('2011_Dist.shp')
shapec
shapec.sort_values('censuscode')
```

```python
Alld=pd.read_csv('clssnumpy.csv')
Alld
```

```python
for_plot = shapec.merge(Alld, left_on = 'censuscode', right_on = 'censuscode');
for_plot.sort_values('censuscode')
```

```python
ax=for_plot.plot(column='censuscode', cmap ='PuBuGn', figsize=(16,10),edgecolor='black',legend ='True');

```

```python
for_plot.plot(column='urban rural ratio', cmap ='PuBuGn', figsize=(16,10),scheme='User_Defined',legend =True,classification_kwds=dict(bins=[0.10,0.3,1.0,2000]));
plt.title('Classification of districts,2011', fontsize=18)
for_plot[for_plot.class.isna()].plot(color='lightgrey', hatch='///', ax=ax)
plt.title('Classification of districts,2011', fontsize=18)
```
