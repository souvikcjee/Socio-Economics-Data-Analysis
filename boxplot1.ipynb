```python

import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
x=[1.5,3.5,5.5]
 
df = pd.read_csv('bxpltks11.csv')
df_t=df.T
 
box = plt.boxplot(df_t, patch_artist=True)
 
colors = ['blue', 'green', 'blue', 'green', 'blue', 'green']
 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    green_patch = mpatches.Patch(color='green', label='KS_PARETO')
blue_patch = mpatches.Patch(color='blue', label='KS_RO')
plt.legend(handles=[green_patch, blue_patch])
labels = ['Population', 'Literacy', 'Non-worker']
plt.xticks(x, labels)

 
plt.show()
```
